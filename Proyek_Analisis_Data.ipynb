## Visualization & Explanatory Analysis

### Pertanyaan 1: Kapan waktu paling populer dalam penyewaan sepeda?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.lineplot(data=hour_df, x='hr', y='cnt', marker='o')
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Jam')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.grid(True)
plt.xticks(range(0, 24))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=day_df.groupby('weekday')['cnt'].mean().reset_index(), x='weekday', y='cnt')
plt.title('Tren Penyewaan Sepeda Berdasarkan Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu (0 = Minggu, 6 = Sabtu)')
plt.ylabel('Rata-rata Jumlah Penyewaan Sepeda')
plt.grid(True)
plt.show()
```

### Pertanyaan 2: Bagaimana pengaruh cuaca terhadap penyewaan sepeda?

```python
hour_df['weathersit'] = hour_df['weathersit'].replace({
    1: 'Clear/Cloudy',
    2: 'Mist/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Rain/Snow'
})

plt.figure(figsize=(10,6))
sns.boxplot(data=hour_df, x='weathersit', y='cnt')
plt.title('Pengaruh Cuaca terhadap Jumlah Penyewaan Sepeda')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.grid(True)
plt.show()
```

- Bagaimana tren penyewaan sepedda berdasarkan hari kerja dan hari libur?

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=hour_df, x='workingday', y='cnt')
plt.title('Tren Penyewaan Sepeda Berdasarkan Hari Kerja dan Hari Libur')
plt.xlabel('Hari Kerja (1 = Ya, 0 = Tidak)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.grid(True)
plt.xticks([0, 1], ['Hari Libur', 'Hari Kerja'])
plt.show()
```

- Bagaimana perbandingan penyewaan sepeda antara pengguna casual dan registered?

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=hour_df, x='hr', y='casual', label='Casual', color='orange')
sns.lineplot(data=hour_df, x='hr', y='registered', label='Registered', color='blue')
plt.title('Perbandingan Penyewaan Sepeda Antara Casual dan Registered Berdasarkan Jam')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.legend(title='Tipe Pengguna')
plt.grid(True)
plt.show()
```

**Insight:**
- Waktu paling populer dalam penyewaan sepeda yaitu pada jam 17.00.
- Hari dengan tren penyewaan tertinggi yaitu pada hari Jumat.
- Jumlah penyewaan sepeda paling tinggi yaitu ketika cuaca cerah atau berawan. Pada cuaca buruk terutama saat hujan lebat atau salju, mengurangi minat orang dalam menyewa sepeda.
- Jumlah penyewaan sepeda rata-rata hampir sama antara hari kerja dan hari libur, namun puncak penyewaan lebih tinggi pada hari kerja.
- Tren penyewaan antara pengguna casual dan registered sangat berbeda.

## Analisis Lanjutan

- RFM Analysis (Recency, Frequency, Monetary) untuk pengguna registered

```python
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

latest_date = hour_df['dteday'].max()
hour_df['Recency'] = (latest_date - hour_df['dteday']).dt.days
frequency = hour_df.groupby('registered')['cnt'].sum().reset_index()
hour_df['Monetary'] = hour_df['cnt'] * hour_df['hr'] 
rfm = hour_df.groupby('registered').agg({
    'Recency': 'min',
    'cnt': 'sum',
    'Monetary': 'sum'
}).reset_index()
rfm.columns = ['User Type', 'Recency (Days)', 'Frequency (Total Rentals)', 'Monetary (Total Time Rented)']
print(rfm)
```

- Clustering

```python
# Mengelompokkan berdasarkan hari kerja (workingday) dan hari (weekday)
hour_df['workingday_cluster'] = hour_df['workingday'].apply(lambda x: 'Working Day' if x == 1 else 'Non-Working Day')

# Group by weekday untuk melihat perbedaan penggunaan pada hari kerja dan libur
day_trend = hour_df.groupby('workingday_cluster')['cnt'].sum().reset_index()
print(day_trend)

```

```python
# Mengelompokkan jam (hr) ke dalam 4 kategori: pagi, siang, sore, malam
hour_df['time_of_day'] = pd.cut(hour_df['hr'], bins=[0, 6, 12, 18, 24], labels=['Malam', 'Pagi', 'Siang', 'Sore'])

# Group by time of day untuk melihat perbedaan penggunaan sepeda di setiap waktu
time_trend = hour_df.groupby('time_of_day')['cnt'].sum().reset_index()
print(time_trend)
```

**Insight:**
- RFM Analysis memberikan gambaran tentang seberapa sering dan kapan pengguna menyewa sepeda serta durasi penggunaannya.
- Clustering dengan manual grouping dan binning membantu unutk memahami perilaku pengguna pada hari kerja dan libur serta pengelompokkan waktu. Dari hasil clustering diketahui bahwa penyewaan lebih banyak pada hari kerja dan waktu paling populer yaitu pada siang hari (pukul 12.00-18.00)
